## ðŸŸ¦ 1. NLP â€” Sentiment Analysis con DistilBERT

```python

from transformers import pipeline

nlp_classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
text = "Artificial intelligence is transforming education."
result = nlp_classifier(text)
result
```

## ðŸŸ© 2. VisiÃ³n â€” Image Classification con ResNet-50

```python

from transformers import AutoImageProcessor, ResNetForImageClassification
from PIL import Image
import requests
import torch

processor = AutoImageProcessor.from_pretrained("microsoft/resnet-50")
model = ResNetForImageClassification.from_pretrained("microsoft/resnet-50")

url = "https://huggingface.co/datasets/mishig/sample_images/resolve/main/tiger.jpg"
image = Image.open(requests.get(url, stream=True).raw)

inputs = processor(images=image, return_tensors="pt")
with torch.no_grad():
    logits = model(**inputs).logits

predicted = logits.argmax(-1).item()
model.config.id2label[predicted]
```

## ðŸŸ§ 3. Audio â€” ASR con Whisper Tiny

```python

from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch
import soundfile as sf

audio, sr = sf.read("audio.wav")

processor = WhisperProcessor.from_pretrained("openai/whisper-tiny")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny")

inputs = processor(audio, sampling_rate=sr, return_tensors="pt")
generated_ids = model.generate(inputs["input_features"])
processor.batch_decode(generated_ids, skip_special_tokens=True)
```
