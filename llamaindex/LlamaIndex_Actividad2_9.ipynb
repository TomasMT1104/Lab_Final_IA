# Actividad 2.9 — LLaMaIndex (RAG)
Este notebook construye un pipeline RAG usando 5 documentos académicos sobre IA en educación.

```

!pip install llama-index==0.10.30 llama-index-readers-file

from llama_index.core import SimpleDirectoryReader, VectorStoreIndex

# Cargar documentos
documents = SimpleDirectoryReader('/mnt/data').load_data()
len(documents)
```

```

# Crear índice vectorial RAG
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# Consulta de prueba
response = query_engine.query("What are the main challenges of implementing AI in education?")
response
```

El resultado anterior demuestra grounding basado en los documentos cargados.
