# **1Ô∏è‚É£ Instalaci√≥n y uso de Modelscope**

> ‚ö†Ô∏è *Importante:*  
Kaggle **NO** permite instalar Modelscope por restricciones de red,  
pero este c√≥digo funciona perfectamente en **Google Colab** o **entorno local con internet**.



```

# ‚úî MODELSCOPE ‚Äî Modelo de similitud sem√°ntica
# Este bloque funciona en Google Colab o PC local con internet.

!pip install modelscope

from modelscope.pipelines import pipeline
from modelscope.utils.constant import Tasks

# Cargar modelo
model = pipeline(Tasks.sentence_similarity, model='damo/sentence_similarity_general')

# Frases de prueba
sentences = {
    "sentence1": "Artificial intelligence is transforming education.",
    "sentence2": "AI is changing how students learn."
}

# Inferencia
result = model(sentences)
print("Modelscope Similarity Score:", result)
```


---

# **2Ô∏è‚É£ Comparaci√≥n con Hugging Face**

Aqu√≠ usamos el modelo equivalente:

üìå `sentence-transformers/paraphrase-MiniLM-L6-v2`



```

# ‚úî HUGGING FACE ‚Äî Modelo de similitud sem√°ntica

!pip install sentence-transformers

from sentence_transformers import SentenceTransformer, util

model_hf = SentenceTransformer('paraphrase-MiniLM-L6-v2')

s1 = "Artificial intelligence is transforming education."
s2 = "AI is changing how students learn."

emb1 = model_hf.encode(s1, convert_to_tensor=True)
emb2 = model_hf.encode(s2, convert_to_tensor=True)

similarity = util.cos_sim(emb1, emb2)
print("HuggingFace Similarity Score:", similarity.item())
```
